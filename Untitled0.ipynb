```python
# 1. Instalação das ferramentas
print("⏳ Instalando bibliotecas... aguarde um minuto.")
!pip install git+https://github.com/openai/whisper.git openai gtts -q
!sudo apt update && sudo apt install ffmpeg -q

# 2. Importações e Conexão com Drive
import whisper
import openai
from gtts import gTTS
import os
from google.colab import userdata, drive

print("📂 Por favor, autorize o acesso ao Google Drive na janela que vai abrir.")
drive.mount('/content/drive')

# 3. Definição da Função de Inteligência
def processar_assistente(caminho_audio):
    try:
        # O Ouvido (Whisper)
        print("🎙️ Whisper ouvindo o arquivo...")
        model = whisper.load_model("base")
        result = model.transcribe(caminho_audio, fp16=False)
        pergunta = result["text"]
        print(f"💬 Você disse: {pergunta}")

        # O Cérebro (GPT)
        print("🧠 GPT pensando na resposta...")
        client = openai.OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": pergunta}]
        )
        resposta_ia = response.choices[0].message.content
        print(f"🤖 IA respondeu: {resposta_ia}")

        # A Voz (gTTS)
        voz = gTTS(resposta_ia, lang='pt')
        voz.save("resposta_final.mp3")
        print("✅ Sucesso! O áudio 'resposta_final.mp3' foi gerado na pasta lateral.")

    except Exception as e:
        print(f"❌ Erro: {e}")

print("🚀 AMBIENTE PRONTO!")
```

```python
# Usando o caminho que confirmamos nas suas capturas anteriores
meu_audio = '/content/drive/MyDrive/Audio overviews/Audio de gisele.mp3'

processar_assistente(meu_audio)
```
